# Assingment 1


```r
df <- read.csv("loan applicants.csv")
head(df)
```

```r
c <- rainbow(2,  start = 0.0399, end = 0.59)
approved_labels <- list("Not Approved", "Approved")
sex_labels <- list("M", "F")
```

## Approval by Mean Income


```r
par(bg="white")
t <- aggregate(Income ~ Approved, df, mean)
print(t)
barplot(t$Income, names.arg=approved_labels, col=c, main="Approval by Mean Income")
```

```r
par(bg="white")
t <- aggregate(Age ~ Approved, df, mean)
print(t)
barplot(t$Age, names.arg=approved_labels, col=c, main="Approval by Mean Age")
```

## Approval by level of Education


```r
par(bg="white")
t <- table(df$Approved, df$Education)
print(t)
barplot(t, beside =TRUE, col=c, main="Approval by level of Education")
```

```r
par(bg="white")
t <- table(df$Approved, df$Employ)
print(t)
barplot(t, beside =TRUE, col=c, main="Approval by level of Employment")
```

## Gender vs Approval


```r
par(bg="white")
t <- table(df$Approved, df$Gender)
print(t)
p <- prop.table(t,margin = 1)
print(p)
barplot(t, beside =FALSE, col=c, main="Gender vs Approval")
```

## Mean Savings by Approval


```r
par(bg="white")
t <- aggregate(Savings ~ Approved, df, mean)
print(t)
barplot(t$Savings,  names.arg=approved_labels, main="Approval by Mean Savings", col=c)
```

```r
t <- prop.table(aggregate(Debt ~ Approved, df, sum))
t
```

```r
par(bg="white")
t <- prop.table(aggregate(Debt ~ Approved, df, sum))
pie(t$Debt, labels=list(sprintf("Not Approved (%f%%)", t[1,]$Debt), 
                sprintf("Approved (%f%%)", t[2,]$Debt)), main="Total Debt by Approval", col=c)
```

```r
par(bg="white")
a <- aggregate(Income ~ Approved, df, mean)
b <- aggregate(Savings ~ Approved, df, mean)
d <- aggregate(Debt ~ Approved, df, mean)
```

```r
par(bg="white")
matrix_data <- matrix(c(a$Income, b$Savings, d$Debt), nrow = 3, byrow = TRUE)
tbl <- as.table(matrix_data)
rownames(tbl) <- c("Income", "Savings", "Debt")
colnames(tbl) <- c(0,1)
print(tbl)
barplot(t(tbl), beside =TRUE, col=c, main="Mean Income vs Savings vs Debt by Approval")
```
