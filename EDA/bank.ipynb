## Whether having a housing loan/any loan have an impact on subscribing to a term deposit.


```r
par(bg="white")
table(housing, y)
prop.table(table(housing, y), 2)
barplot(prop.table(t(table(housing, y)), 2), col=colors, legend=sub_c)
```

```r
par(bg="white")
l = "housing: "
barplot(table(housing, y), col=c("red", "yellow", "green"), legend=c(sprintf("%s: no", l), sprintf("%s: unknown", l), sprintf("%s: yes", l)), beside=TRUE, main="Housing loan vs term deposit")
```

```r
par(bg="white")
l = "loan: "
barplot(table(loan, y), col=c("red", "yellow", "green"), legend=c(sprintf("%s: no", l), sprintf("%s: unknown", l), sprintf("%s: yes", l)), beside=TRUE, main="Personal loan vs term deposit")
```
